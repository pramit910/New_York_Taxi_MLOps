### Question 2 - Computing the duration

```python
# duration of a ride -> (tpep_dropoff_datetime - tpep_pickup_datetime)

def convert_to_minutes(time):
    return time.total_seconds()/60
    

# df.apply(lambda x: (df_jan['tpep_dropoff_datetime'] - df_jan['tpep_pickup_datetime']).total_seconds()/60, axis = 1)
df_jan['duration'] = (df_jan['tpep_dropoff_datetime'] - df_jan['tpep_pickup_datetime'])
df_jan['duration'] = df_jan['duration'].apply(convert_to_minutes) 
```
